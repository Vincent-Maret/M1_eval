## Analyse des facteurs de réussite scolaire d'une classe de mathématiques
Source: [kaggle.com](https://www.kaggle.com/uciml/student-alcohol-consumption)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('student-mat.csv')
data.head()
```

```python
data.info()
```

## Description du dataset

Le dataset contient 395 lignes correspondant chacune à un étudiant en mathématique dans le secondaire. Pour chaque étudiant, nous avons un jeu de 33 donnée (colonnes).

- **school**: école de l'étudiant. *GP* (Gabriel Pereira) ou *MS* (Mousinho da Silveira) **binary**
- **sex**:  sexe de l'étudiant, *F* (female) ou *M* (male) **binary**
- **age**: age de l'étudiant **numeric**
- **address**: type de logement, *U* (urban) ou *R* (rural) **binary**
- **famsize**: taille de la famille, *LE3* (moins ou 3 membres) ou *GT3* (plus de 3 membres) **binary**
- **Pstatus**: statut de cohabitation des parents, *T* (vivent ensemble) ou *A* (vivent séparément) **binary**
- **Medu**: niveau d'éducation de la mère. *0* (pas d'éducation), *1* (4th grade/primaire), *2* (5th to 9th grade/collège), *3* (secondary/lycée) et *4* (higher education/supérieur) **numeric**
- **Fedu**: niveau d'éducation du père. **numeric** 
- **Mjob**: travail de la mère. *teacher*, *health*, *services*, *at_home* ou *other* **string**
- **Fjob**: travail du père **string**
- **reason**: raison du choix de l'école *home* (près de la maison), *reputation* (réputation de l'école) *course* (préférence des cours) ou *other* **string**
- **guardian**: tuteur légal, *mother*, *father* ou *other* **string**
- **traveltime**: temps de trajet école - maison en heure. **numeric**
- **studytime**: temps d'étude par semaine. 1 = 10h **numeric** 
- **failures**: nombre de classe redoublés. Maximum 4 **numeric**
- **schoolsup**: extra support scolaire *yes* ou *no* **boolean**
- **famsup**: support scolaire apporté par la famille, *yes* ou *no* **boolean*
- **paid**: extra support scolaire payant en math, *yes* ou *no* **boolean**
- **activities**: activités extra scolaire *yes* ou *no* **boolean**
- **nursery**: aide au devoir/étude *yes* ou *no* **boolean**
- **higher**: veut entreprendre des études supérieurs *yes* ou *no* **boolean**
- **internet**: accès internet à la maison *yes* ou *no* **boolean**
- **romantic**: relation amoureuse *yes* ou *no* **boolean**
- **famrel**: qualité des relations familliales, de *1* (bas) à *5* (haut) **numeric**
- **freetime**: temps libre après l'école, de *1* (bas) à *5* (haut) **numeric**
- **goout**: taux de sortie avec des amis, de *1* (bas) à *5* (haut) **numeric**
- **Dalc**: taux de consomation d'acool un jour de semaine, de *1* (bas) à *5* (haut) **numeric**
- **Walc**: taux de consomation d'alcool le weekend, de *1* (bas) à *5* (haut) **numeric**
- **health**: niveau global de santé, de *1* (bas) à *5* (haut) **numeric**
- **absences**: nombre d'absences à l'école, de *0* à *93* **numeric**
- **G1**: moyenne au 1er trimestre, de *0* à *20* **numeric**
- **G2**: moyenne au 2nd trimestre, de *0* à *20* **numeric**
- **G3**: moyenne au 3ème trimestre, de *0* à *20* **numeric**

On peut diviser ces variables en plusieurs catégories:
- **La famille**: famsize, Pstatus, Medu, Fedu, Mjob, Fjob, guardian, famsup, famrel
- **Le soutient scolaire**: schoolsup, famsup, paid, nursery
- **Le "life style"**: activities, romantic, freetime, goout, Dalc, Walc
- **L'environement**: address, traveltime, internet, school
- **L'assiduité scolaire**: studytime, higher, absences, failures
- **Caractéristiques de l'étudiant**: sex, age, health

Ici nous créons la colonne fullGrades corerspondant à la moyenne annuelle pour simplifier l'analyse.

```python
data = data.assign( fullGrades = (data["G1"] + data["G2"] + data["G3"]) / 3 )
```

```python
plt.figure(figsize=(16,8))
plt.boxplot(data.fullGrades, showmeans=True);
```

## Analyse des variables numériques corélatives à la réussite scolaire

```python
plt.figure(figsize=(14,8))
sns.heatmap(data.corr().sort_values(by=['fullGrades'],ascending=False),annot=True);
```

On constate que le niveau d'éducation des parents, le temps passé à étudier et la qualité des relations familiales sont les variables impactant le plus la réussite scolaire. La famille semble donc impacter le plus la réussite scolaire.

## La famille

```python
sns.catplot(x="Medu", y="fullGrades", kind="box", data=data, height=4)
plt.xlabel("Niveau d'éducation de la mère")
plt.ylabel("Moyenne annuelle")
sns.catplot(x="Fedu", y="fullGrades", kind="box", data=data, height=4)
plt.xlabel("Niveau d'éducation du père")
plt.ylabel("Moyenne annuelle");
```

Globalement, les enfants dont les parents n'ont pas d'éducation obtiennent de meilleur résultats à l'école. C'est aussi le groupe le plus homogène.
Cependant, à partir du niveau d'éducation 1 (primaire) jusqu'au 4 (supérieur), la tendance s'inverse, et plus on vat vers un niveau d'édcation supérieur, plus les résultats sonthaut. Ce sont aussi les groupes les plus dispersés.

```python
sns.catplot(x="famrel", y="fullGrades", kind="box", data=data, height=4)
plt.xlabel("Relations familiales")
plt.ylabel("Moyenne annuelle");
```

La qualité des relations familiales semble avoir un impact. Plus elles sont bonnes, plus la moyenne globale semble dispersée.

## Le soutient scolaire

```python
sns.catplot(x="schoolsup", y="fullGrades", kind="box", data=data, height=4)
plt.xlabel("Support scolaire")
plt.ylabel("Moyenne annuelle")
sns.catplot(x="paid", y="fullGrades", kind="box", data=data, height=4)
plt.xlabel("Support scolaire payant")
plt.ylabel("Moyenne annuelle")
sns.catplot(x="famsup", y="fullGrades", kind="box", data=data, height=4)
plt.xlabel("Sopport scolaire apporté par la famille")
plt.ylabel("Moyenne annuelle")
sns.catplot(x="nursery", y="fullGrades", kind="box", data=data, height=4)
plt.xlabel("Etude")
plt.ylabel("Moyenne annuelle");
```

Le support scolaire ne semble pas impacté sur la réussite scolaire. Cependant on peut observer que ceux qui en bénificient ont globalement une moyenne plus basse que ceux qui nen bénificient pas.

### Evolution des notes sur l'année

#### Support scolaire

```python
sns.catplot(x="schoolsup", y="G1", kind="box", data=data, height=4)
plt.xlabel("Support scolaire")
plt.ylabel("Moyenne annuelle")
sns.catplot(x="schoolsup", y="G2", kind="box", data=data, height=4)
plt.xlabel("Support scolaire")
plt.ylabel("Moyenne annuelle")
sns.catplot(x="schoolsup", y="G3", kind="box", data=data, height=4)
plt.xlabel("Support scolaire")
plt.ylabel("Moyenne annuelle");
```

#### Support scolaire payant

```python
sns.catplot(x="paid", y="G1", kind="box", data=data, height=4)
plt.xlabel("Support scolaire")
plt.ylabel("Moyenne annuelle")
sns.catplot(x="paid", y="G2", kind="box", data=data, height=4)
plt.xlabel("Support scolaire")
plt.ylabel("Moyenne annuelle")
sns.catplot(x="paid", y="G3", kind="box", data=data, height=4)
plt.xlabel("Support scolaire")
plt.ylabel("Moyenne annuelle");
```

Un support scolaire gratuit ou payant ne semble pas impacter l'évolution des résultats au cours de l'année.

## Le "life style"

### Relation de couple

```python
sns.catplot(x="romantic", y="fullGrades", kind="box", data=data, height=4)
plt.xlabel("Relation de couple")
plt.ylabel("Moyenne annuelle");
```

### Temps Libre

```python
sns.catplot(x="freetime", y="fullGrades", kind="box", data=data, height=4)
plt.xlabel("Temps libre")
plt.ylabel("Moyenne annuelle");

sns.catplot(x="goout", y="freetime", kind="box", data=data, height=4)
plt.xlabel("Sorties")
plt.ylabel("Moyenne annuelle");
```

```python
sns.lineplot(x='goout', y='freetime', data=data);
```

Le temps libre et le fait de sortir semblent avoir un impact positif sur la réussite scolaire.

### Consomation d'alcool le weekend vs en semaine

```python
plt.figure(figsize=(16,8))
plt.subplot(121)
plt.boxplot(data.Walc, showmeans=True)
plt.title('Répartition de la consomation d\'alcool le weekend')
plt.subplot(122)
plt.boxplot(data.Dalc, showmeans=True)
plt.title('Répartition de la consomation d\'alcool en semaine');
```

Globalement, les étudiants consomment beaucoup plus d\'alcool le weekend qu'en semaine.

**En semaine**: la médiane se trouve au niveau le plus bas (1). La grande majorité des étudiants ne consomment donc pas d'alcool en semaine. L'écart entre la medianne et la moyenne est de 0.5. Il est deux fois plus grand que le weekend.La répartition de consommation est donc moins homogène en semaine, et cela se confirme par la présence de deux individus abbérants, avec une consomation de 4 et 5 en semaine.

**Le weekend**: avec une médianne de 2, on constate que les étudiants consomment plus. L'écart entre la médiane et la moyenne est de 0.25, la répartition est donc plus homogène qu'en semaine. Ceci dit, la consomation maximale (5) n'est plus une abbération mais un plafond. Cela ne décrivant pas une majorité, la consomation maximale globale étant à 3, soit 1 de plus qu'en semaine. La consomation minimale (1) reste dans la normalité.

## Incidence de la consomation d'alcool sur la réussite scolaire

```python
plt.figure(figsize=(10,5))
plt.bar(data.Walc, data.fullGrades, label='Weekend')
plt.bar(data.Dalc, data.fullGrades, label='Semaine')
plt.xlabel('Consommation d\'alcool')
plt.ylabel("Moyenne annuelle")
plt.legend();
```

On peut constater que la consommation d'alcool a un effet négatif bien plus important en semaine que le weekend. On observe une tendance à la baisse le weekend mais c'elle ci est beaucoup plus importante en semaine.

```python
sns.lineplot(x='goout', y='Walc', data=data);
```

```python
sns.lineplot(x='goout', y='Dalc', data=data);
```

La consommation d'alcool évolue avec le taux de sortie. On a vu que le taux de sortie à un impact positif sur la réussite scolaire mais que la consomation d'alcool en semaine avait un impact négatif sur la scolarité.
